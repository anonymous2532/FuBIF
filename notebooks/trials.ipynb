```python



import pandas as pd
import pickle
import numpy as np

with open("/Yourpath/FuBIF/results/numeric/performances/2024-05-28_12-09-28.pkl", "rb") as f:
    data_1 = pickle.load(f)

with open("/Yourpath/FuBIF/results/numeric/performances/Paraboloid_2024-06-14_03-05-16.pkl", "rb") as f:
    data_2 = pickle.load(f)

with open("/Yourpath/FuBIF/results/numeric/performances/Ellipsoid_2024-06-14_03-05-41.pkl", "rb") as f:
    data_3 = pickle.load(f)
    
with open("/Yourpath/FuBIF/results/numeric/performances/PE2024-06-13_19-12-27.pkl", "rb") as f:
    data_4 = pickle.load(f)   

with open("/Yourpath/FuBIF/results/numeric/performances/2024-06-13_14-30-05.pkl", "rb") as f:
    data_5 = pickle.load(f)    
    
data = {**data_1, **data_2, **data_3, **data_4,**data_5}
data = {"SingleDimensionHyperplane":data["SingleDimensionHyperplane"],
                          "Hyperplane":data["Hyperplane"],
                          "Circle":data["CircleSplitting"], 
                          "Ellipsoid":data["Ellipsoid"], 
                          "Paraboloid":data["Paraboloid"], 
                          "Hyperbolic":data["HyperbolicSplitting"],
                          "Conic" : data["ConicSplitting"],
                          "NN": data["NN"],
                          'IForest': data['IForest'], 
                          'AutoEncoder': data['AutoEncoder'], 
                          'DIF': data['DIF'],}
df = pd.DataFrame(data)
df = pd.json_normalize(data, sep='/').T
df.iloc[0, 0] 
df.index = pd.MultiIndex.from_tuples(df.index.str.split('/').map(tuple))
df = df.unstack(level=1).droplevel(0, axis=1).sort_index(axis=1)

df= df.unstack(level=0)
df=df.T
def apply_mean(cell):
    return np.mean(cell)

# Apply the function to each cell of the DataFrame
df_mean = df.applymap(np.mean)
```

```python
import sys
sys.path.append("../")


import pandas as pd
import pickle
import numpy as np

with open("/Yourpath/FuBIF/results/numeric/performances/NN_2024-06-04_19-51-54.pkl", "rb") as f:
    data = pickle.load(f)
df = pd.DataFrame(data)
df = pd.json_normalize(data, sep='/').T
df.iloc[0, 0] 
df.index = pd.MultiIndex.from_tuples(df.index.str.split('/').map(tuple))
df = df.unstack(level=1).droplevel(0, axis=1).sort_index(axis=1)

df= df.unstack(level=0)
df=df.T
def apply_mean(cell):
    return np.mean(cell)

# Apply the function to each cell of the DataFrame
df_mean_NN = df.applymap(np.mean)
```

```python
df_mean_NN
```

```python
df1_reset = df_mean_dif.reset_index(level=1)
df2_reset = df_mean_NN.reset_index(level=1)

# Concatenate the DataFrames
concat_df = pd.concat([df1_reset, df2_reset], axis=0)

# Sort the resulting DataFrame by index to ensure proper order
concat_df = concat_df.sort_index()

print("Concatenated DataFrame:")
concat_df

```
